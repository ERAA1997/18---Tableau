```python
#import needed modules

import pandas as pd
from datetime import datetime, time
```

```python
#Read in first dataset
df_june = pd.read_csv("Resources/JC-202306-citibike-tripdata-JUNE.csv")
```

```python
#check headers and data
df_june.head()
```

```python
#Read in second dataset
df_dec = pd.read_csv("Resources/JC-202312-citibike-tripdata-DEC.csv")
```

```python
#check headers and data
df_dec.head()
```

```python
#Concatenate both datasets into one
concat_df = pd.concat([df_june,df_dec], axis=0)
```

```python
#check headers and data
concat_df.head()
```

```python
Drop any rows containing Null values
c_c_df = concat_df.dropna()

print(len(c_c_df))
```

```python
#Check datatypes
c_c_df.dtypes
```

```python
#convert columns to datatime to use for creating date and time columns
c_c_df['started_at'] = pd.to_datetime(c_c_df['started_at'])

c_c_df['started at date'] = c_c_df['started_at'].dt.date

c_c_df['started at time'] = c_c_df['started_at'].dt.time

c_c_df['ended_at'] = pd.to_datetime(c_c_df['ended_at'])

c_c_df['ended at date'] = c_c_df['ended_at'].dt.date

c_c_df['ended at time'] = c_c_df['ended_at'].dt.time



```

```python
#Mispelled a heading leading to a randm column, this code used to deleted it out of the df
del c_c_df['endedat time']
```

```python
#check headers and data
c_c_df.head()
```

```python
#Check datatypes again
c_c_df.dtypes
```

```python
# Calculate the time difference
c_c_df['started_at'] = pd.to_datetime(c_c_df['started_at'])
c_c_df['ended_at'] = pd.to_datetime(c_c_df['ended_at'])

 
c_c_df['duration'] = (c_c_df['ended_at'] - c_c_df['started_at']).dt.total_seconds() / 60
```

```python
#Check headers and data
c_c_df.head()
```

```python
#Export CSV for usage
c_c_df.to_csv('cleaned_bike_data.csv')
```
