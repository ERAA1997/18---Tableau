```python
c_c_df = concat_df.dropna()

print(len(c_c_df))
```

```python
c_c_df.dtypes
```

```python
c_c_df['started_at'] = pd.to_datetime(c_c_df['started_at'])

c_c_df['started at date'] = c_c_df['started_at'].dt.date

c_c_df['started at time'] = c_c_df['started_at'].dt.time

c_c_df['ended_at'] = pd.to_datetime(c_c_df['ended_at'])

c_c_df['ended at date'] = c_c_df['ended_at'].dt.date

c_c_df['ended at time'] = c_c_df['ended_at'].dt.time



```

```python
del c_c_df['endedat time']
```

```python
c_c_df.head()
```

```python
c_c_df.dtypes
```

```python
c_c_df['started at time'] = c_c_df['started at time'].apply(lambda x: datetime.combine(datetime.min, x) - datetime.min)

c_c_df['ended at time'] = c_c_df['ended at time'].apply(lambda x: datetime.combine(datetime.min, x) - datetime.min)

c_c_df['Trip Length'] = (c_c_df['ended at time'] - c_c_df['started at time']).dt.total_seconds() / 60
```

```python
c_c_df['started_at'] = pd.to_datetime(c_c_df['started_at'])
c_c_df['ended_at'] = pd.to_datetime(c_c_df['ended_at'])

# Calculate the time difference between 'ended_at' and 'started_at'
c_c_df['duration'] = (c_c_df['ended_at'] - c_c_df['started_at']).dt.total_seconds() / 60
```
